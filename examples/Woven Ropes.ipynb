## Woven ropes

From

> Frank A. Farris & Nils Kristian Rossing (1999) Woven Rope Friezes, *Mathematics Magazine*, 72:1, 32-38, DOI: [10.1080/0025570X.1999.11996695](https://doi.org/10.1080/0025570X.1999.11996695)


```python
import numpy as np
from numpy import pi, sin, cos, exp
import matplotlib.pyplot as plt
```

Represent a woven curve as a 'decorative' component with period $2\pi$:

$$
m(t) = \left(
\sum_{n=0}^{\infty} a_n \sin{\frac{n \pi t}{L} } + b_n \cos{\frac{n \pi t}{L} },
\sum_{m=0}^{\infty} c_m \sin{\frac{m \pi t}{L} } + d_m \cos{\frac{m \pi t}{L} },
\right)
$$

combined with a translational component:

$$
c(t) = m(t) + \left( \frac{K t}{2 L}, 0 \right)
$$

Symmetry is determined by whether the $x(t)$ and $y(t)$ components of $c(t)$ can be decomposed into even/odd or glide-positive/negative components:

$$
x(t) = x_e(t) + x_o(t) = \frac{x(t)+x(-t)}{2} + \frac{x(t)-x(-t)}{2}
$$
$$
x(t) = x_{G+}(t) + x_{g-}(t) = \frac{x(t)+x(t+L)}{2} + \frac{x(t)-x(t+L)}{2}
$$

|Symmetry|Strands|Conditions|Recipe|
|---|---|---|---|
|$G_1$ p111| 1 | none| none|
|$G_2$ p211| 1 | $x_e$ = $y_e$ = 0 | sin only|
|$G_3$ p1m1| 2 | $G_1$ or $G_7$ in mirrored strands | none|
|$G_4$ p11m| 1 | $x_e = 0$, $y_o = 0$ | sin in x, cos in y|
|$G_5$ p2mm| 2 | $G_2$, $G_4$, or $G_6$ in mirrored strands|
|$G_6$ p2mg| 1 | $X_e = y_e = x_{g-}=y_{g+}$ = 0|$n$ even, $m$ odd, sin only|
|$G_7$ p11g| 1 | $ x_{g-}=y_{g+} = 0$|$n$ even, $m$ odd, sin and cos|


```python
# Fig 1, p211
t = np.linspace(0, 10 * pi, 1000)
x = 9 * t / (2 * pi) - sin(2 * t) - 5 * sin(3 * t) - 5 * sin(4 * t)
y = sin(t) + 2 * sin(2 * t)

plt.plot(x, y)
```

```python
# Fig 4 (fixed), p211
t = np.linspace(0, 13 * pi, 1000)
x = 0.9 * t / (2 * pi) + sin(2 * t)
y = -0.8 * cos(t)

plt.plot(x, y)
```

```python
# Fig 5, p2mg
plt.figure(figsize=(14, 3))
t = np.linspace(0, 6 * pi, 1000)
x = 13 * t / (2 * pi) - 1.5 * sin(2 * t) - 3.5 * sin(4 * t)
y = -0.3 * sin(t) - 3.2 * sin(3 * t)

plt.plot(x, y, "b-", x, -y, "g-")
```
